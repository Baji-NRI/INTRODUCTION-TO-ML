```python

# loops -> sub concept
# conditions
# expressions
# statements
# if - else
# syntax : if(condition): EV elif(condition): EV else: EV
a = 10
b = 20
c = 300
if (a>b):
   print(f"{a} is greater than {b}")
elif (a<b<c):
   print(f"{c} is greater than {a} and {b}")
else:
   print(f"{b} is greater than {a}")
```
